```python
df = pd.read_csv('Salary_dataset.csv')
```

```python
df.corr()
```

```python
X = df['YearsExperience']
y = df['Salary']  # Altere isso se a variável alvo for diferente

plt.plot(X, y, "b.")
plt.show()
```

```python
X_b = np.c_[np.ones((30, 1)), X]
theta_param = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
X_new = np.array([[0], [11]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_param)

plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.show()
```

```python
# Prevendo os valores para todo o conjunto original X
y_pred = X_b.dot(theta_param)  # Use X_b para prever todos os pontos

# Calculando o Erro Quadrático Médio (MSE)
mse = np.mean((y - y_pred) ** 2)

print("Erro Quadrático Médio (MSE):", mse)
```

```python
# Calculando o coeficiente de determinação R^2
ss_tot = np.sum((y - np.mean(y)) ** 2)
ss_res = np.sum((y - y_pred) ** 2)
r2_score = 1 - (ss_res / ss_tot)

print("Coeficiente de Determinação R^2:", r2_score)
```
