```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Introduza os valores desejados para os pontos xi e yi

xi = np.array([1.2, 1.4, 1.6, 2.1, 2.3, 3.0, 3.1, 3.3, 3.3, 3.8, 4.0, 4.1, 4.1, 4.2,
                  4.6, 5.0, 5.2, 5.4, 6.0, 6.1, 6.9, 7.2, 8.0, 8.3, 8.8, 9.1, 9.6, 9.7,
                  10.4, 10.6], dtype='double')
yi = np.array([39344, 46206, 37732, 43526, 39892, 56643, 60151, 54446, 64446,
                  57190, 63219, 55795, 56958, 57082, 61112, 67939, 66030, 83089,
                  81364, 93941, 91739, 98274, 101303, 113813, 109432, 105583,
                  116970, 112636, 122392, 121873], dtype='double')
```

```python
n = len(xi)
```

```python
sum_xi = np.sum(xi)
sum_xi2 = np.sum(xi**2)
sum_yi = np.sum(yi)
sum_xy = np.sum(xi*yi)
```

```python
a = (n*sum_xy - sum_xi*sum_yi)/(n*sum_xi2 - sum_xi**2)
```

```python
b = (sum_yi - a*sum_xi)/n
```

```python
print('y =', a, 'x + ', b)
```

```python
x = input('Digite o valor de x dentre o intervalo: ')
x = float(x)
```

```python
fx = a*x + b
fx
```

```python
result = a* xi + b
print(result)
```

```python
# Subtraindo termo a termo
desvio = [(v1 - v2)**2 for v1, v2 in zip(yi, result)]

print(desvio)

print(sum(desvio))
```

```python
# Plotando os pontos e a linha de regressão
plt.figure(figsize=(8, 6))
plt.plot(xi, yi, 'bo', label='Pontos (xi, yi)')  # Pontos em azul
plt.plot(xi, result, 'r-', label=f'y = {a:.2f}x + {b:.2f}')  # Linha de regressão em vermelho
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste Linear')
plt.legend()
plt.grid(True)
plt.show()
```
