```python
import numpy as np
import matplotlib.pyplot as plt

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
X_b = np.c_[np.ones((100, 1)), X]
theta_param = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python

X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_param)

plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

```python

from sklearn.linear_model import LinearRegression

linear_reg = LinearRegression()
linear_reg.fit(X, y)
linear_reg.intercept_ #coeficiente linear
linear_reg.coef_ #coeficiente angular
linear_reg.predict(X_new)
y_predict = linear_reg.predict(X_new)
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
# Prevendo os valores para todo o conjunto original X
y_pred = X_b.dot(theta_param)  # Use X_b para prever todos os pontos

# Calculando o Erro Quadrtico Mdio (MSE)
mse = np.mean((y - y_pred) ** 2)

print("Erro Quadrtico Mdio (MSE):", mse)
```

```python
# Calculando o coeficiente de determinao R^2
ss_tot = np.sum((y - np.mean(y)) ** 2)
desvio = np.sum((y - y_pred) ** 2)
r2_score = 1 - (desvio / ss_tot)

print("Coeficiente de Determinao R^2:", r2_score)
print("Desvio:", desvio)
```
